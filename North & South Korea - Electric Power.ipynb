```python
import pandas
from matplotlib import pyplot
from matplotlib import font_manager, rc

font = 'malgun.ttf' # Very popular Korean font
font_name = font_manager.FontProperties(fname = font).get_name()
rc('font', family = font_name)
```

```python
df = pandas.read_excel('data/남북한_발전전력량.xlsx')
df
df['남북한별'] = df['남북한별'].ffill() 
df2 = df.loc[[0, 3], '1990':] 
df2.index = ['north', 'south']
df2
```

```python
df2.loc['south'].plot(kind = 'line')
```

```python
north = df2.loc['north']
```

```python
north = df2.loc['north']
north.plot(x = north.index, y = north.values, figsize = (11, 4))
```

```python
df2_T = df2.transpose() 
df2_T.plot(figsize = (11, 4))
```

```python
df2_T.columns = ['북한', '남한']
df2_T.plot(figsize = (11, 4))
pyplot.title('남북한 발전전력량(1900-2020)', size = 20)
pyplot.xlabel('기간(year)', size = 14)
pyplot.ylabel('전력량(억kwh)', size = 14)
pyplot.xticks(rotation = 45)
pyplot.ylim(0, 8000)
pyplot.legend(loc = 'best')
pyplot.show() 
```
